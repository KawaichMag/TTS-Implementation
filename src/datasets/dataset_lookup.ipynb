```python
from pathlib import Path

DATASET_PATH = Path("./golden_set")

audio_path = next(DATASET_PATH.iterdir())
audio_path
```

```python
import librosa

audio, sr = librosa.load(audio_path)

audio
```

```python
import matplotlib.pyplot as plt

plt.plot(audio)
plt.show()
```

```python
import numpy as np

batched_audio = librosa.stft(audio)

fig, ax = plt.subplots()
spec = librosa.display.specshow(
    librosa.amplitude_to_db(np.abs(batched_audio), ref=np.max),
    y_axis="mel",
    x_axis="time",
    ax=ax,
)
print(batched_audio.shape)

fig.colorbar(spec, ax=ax)
plt.show()
```
